```python
import pandas as pd
from mplsoccer.pitch import Pitch
```

```python
# read in csv
df = pd.read_csv('C:\\Users\\Administrator\\Desktop\\Data_Analysis\\matchevent_BARvsVAL.csv')
```

```python
df
```

```python
df = df[df['teamId'] == 'Barcelona']
df
```

```python
# Create a new column
df['passer'] = df['playerId']
df['recipient'] = df['playerId'].shift(-1)

# Create a new df
passes = df[df['type'] == 'Pass']
successful = passes[passes['outcome'] == 'Successful']
successful
```

```python
subs = df[df['type'] == 'SubstitutionOff']
subs = subs['minute']
firstSub = subs.min()
firstSub
```

```python
successful = successful[successful['minute'] < firstSub]
successful
```

```python
# Convert to integer
pas = pd.to_numeric(successful['passer'], downcast = 'integer')
rec = pd.to_numeric(successful['recipient'], downcast = 'integer')

successful['passer'] = pas
successful['recipient'] = rec
successful
```

```python
# Find the average locations of x, y coord.             # Find the count of successful passes
average_locations = successful.groupby('passer').agg({'x' : ['mean'], 'y' : ['mean', 'count']})
average_locations.columns = ['x', 'y', 'count']
average_locations
```

```python
pass_between = successful.groupby(['passer', 'recipient']).id.count().reset_index()
pass_between.rename({'id' : 'pass_count'}, axis='columns', inplace=True)

# Merge the average locations df with passer and recipient
pass_between = pass_between.merge(average_locations, left_on='passer', right_index=True)
pass_between = pass_between.merge(average_locations, left_on='recipient', right_index=True, suffixes=['', '_end'])
pass_between
```

```python
# Display passes between players > 3
pass_between = pass_between[pass_between['pass_count'] > 3]
pass_between
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#06D6A0', line_color='#AF125A')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor("#06D6A0")

for idx, row in average_locations.iterrows():
    pitch.annotate(int(idx), xy=(row.x * 1.2, row.y * 0.8),ax=ax,color='white', va='center', ha='center')

# Display arrows for passes
arrows = pitch.arrows(1.2 * pass_between.x, 0.8 * pass_between.y, 
                      1.2 * pass_between.x_end, 0.8 * pass_between.y_end, ax=ax,
                      width=5, headwidth=3, color='#000F08', zorder=1, alpha=0.9)

# Display nodes as players
nodes = pitch.scatter(1.2 * average_locations.x, 0.8 * average_locations.y,
                      s=700, color='#92140C', edgecolors='#9E788F',
                      linewidth=2.5, alpha=1, zorder=1, ax=ax)
```
